```
df_traffic_threshold = df_traffic.max()[1:] + df_traffic.describe().iloc[5,:]
df_traffic_threshold = np.array(df_traffic_threshold)
df_traffic_threshold
```

```
df_addresses['Traffic Threshold'] = df_traffic_threshold
```

```
import random
```

```
def getYears(startYear,endYear):

  years = []
  for i in range(100):
    years.append(random.randint(startYear,endYear))

  return np.array(years)
  
```

```
getYears(2000,2015)
```

```
df_addresses['Construction Year'] = getYears(2000,2015)
df_addresses
```
